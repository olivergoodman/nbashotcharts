```python
import matplotlib.pyplot as plt
import pandas as pandas
import nbashots as nba
```

```python
# GET INDIVIDUAL PLAYER DATA
curry_id = nba.get_player_id("Curry, Stephen")[0]
curry_logs = nba.PlayerLog(curry_id)
curry_logs_df = curry_logs.get_game_logs() #data frame for curry logs
curry_logs_df.head()
```

```python
# GET TEAM DATA
warriors_id = nba.get_team_id("Warriors")
warriors_logs = nba.TeamLog(warriors_id)
warriors_logs_df = warriors_logs.get_game_logs()
warriors_logs_df.head()
```

```python
all_data = nba.get_all_player_ids("all_data")
all_data.head()
```

```python
# USING SHOTS CLASS
shots = nba.Shots()
current_season_shots = shots.get_shots() #every shot taken for current season
current_season_shots.head()

last_season_shots = nba.Shots(season="2014-15").get_shots()
last_season_shots.head()
```

```python
# MAKING SHOT CHARTS

curry_shots_df = nba.Shots(curry_id).get_shots() #get steph's shots

plt.rcParams['figure.figsize'] = (12, 11) #setting size for plots
nba.shot_chart(curry_shots_df.LOC_X, curry_shots_df.LOC_Y,
              title="Stephen Curry FGA 15-16 Season")
plt.show()
```

```python
# TEAM SHOT CHARTS
warriors_shots = nba.Shots(team_id=warriors_id) #create Shots object
warriors_shots_df = warriors_shots.get_shots() #get dubs shots dataframe
warriors_shots_df.head()
nba.shot_chart(warriors_shots_df.LOC_X, warriors_shots_df.LOC_Y,
              title="Warriors FGA 2015-16 Season")
plt.show()
```

```python
# KDE SHOT CHARTS
nba.shot_chart(curry_shots_df.LOC_X, curry_shots_df.LOC_Y,
              kind="kde", title="Stephen Curry FGA 15-16 Season")
plt.show()
```

```python
# HEX SHOT CHARTS
# get a colormap for the KDE shot chart
cmap=plt.cm.gist_heat_r

nba.shot_chart(curry_shots_df.LOC_X, curry_shots_df.LOC_Y,
              kind="hex", cmap=cmap, gridsize=26, title="Stephen Curry FGA 15-16 Season")
plt.show()
```
